 1. What is hypothesis testing in statistics.

- Hypothesis testing is a method in statistics used to make decisions or inferences about population parameters based on sample data. Think of it as a formal way of testing assumptions using evidence gathered from experiments or observations.
- Null Hypothesis (H_0): The default assumption—usually that there's no effect or no difference. It's what we try to disprove.
- Alternative Hypothesis (H_1 or H_a): The rival claim that suggests there is an effect or a difference.
- Test Statistic: A value calculated from the sample that helps us decide whether to reject H_0.
- P-value: The probability of observing results at least as extreme as those in your sample, assuming H_0 is true. A small p-value means the sample data is unlikely under H_0.
- Significance Level (\alpha): A threshold (commonly 0.05) set by the researcher to decide when to reject H_0. If p-value < \alpha, H_0 is rejected.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis.

- Null Hypothesis (H_0)
- This is the default or baseline assumption.
- It states that there is no effect, no difference, or no relationship between variables.
- You're essentially saying: “Things are working as usual.”
- Example: If you’re testing a new drug, H_0 might say the drug has no impact compared to the existing one.


3. What is the significance level in hypothesis testing, and why is it important.

- Significance Level (\alpha) is a threshold we set in hypothesis testing to determine how strong the evidence must be before we reject the null hypothesis. It's a bit like setting the rules before playing the game.
Why It's Important
- Controls the risk of false positives: A lower \alpha means you're being more cautious in declaring an effect.
- Sets the benchmark for your p-value: You only reject H_0 if the p-value < \alpha.
- It defines what “statistically significant” means in your analysis.



4. What does a P-value represent in hypothesis testing.

- A p-value is a critical measure in hypothesis testing that helps you determine how compatible your data is with the null hypothesis. It’s the probability of getting your observed results—or something more extreme—if the null hypothesis were true.


5. How do you interpret the P-value in hypothesis testing.

 - Interpreting the p-value in hypothesis testing boils down to understanding what the number is telling you about your evidence—relative to the assumption that the null hypothesis (H_0) is true.


- It tells you the likelihood of observing your sample data—or something more extreme—if H_0 were true.
- A small p-value means that your results are rare under the null hypothesis, making it questionable.
- A large p-value suggests your results are not surprising, so there's little reason to doubt H_0.


6. What are Type 1 and Type 2 errors in hypothesis testing.

- In hypothesis testing, Type I and Type II errors are two kinds of mistakes you can make when interpreting results—like misjudging a verdict in court.

 Type I Error (False Positive)
- You reject the null hypothesis when it’s actually true.
- It’s like sounding the alarm when there's no real danger.
- Think: diagnosing a healthy person as sick.
 Example:
You test a new medicine.
- H_0: It doesn’t work.
- You reject H_0 based on your sample—but in reality, it doesn’t work.
- You’ve made a Type I error.
-  Probability of making this error = Significance level (\alpha), often set at 0.05

 Type II Error (False Negative)
- You fail to reject the null hypothesis when it's actually false.
- You miss something that’s actually there.
- Think: failing to diagnose a sick person.





7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing.

- In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the research question and what kind of difference you're looking to detect.

 One-Tailed Test
- Tests for an effect in one direction only—either greater than or less than.
- Used when the hypothesis specifies a directional claim.
 Example:
You're testing if a new drug performs better than the existing one:
- H_0: New drug is not better than the old one.
- H_1: New drug is better than the old one.
- You only care if it improves—not if it performs worse.

 Two-Tailed Test
- Tests for an effect in both directions—could be greater or less.
- Used when the hypothesis is non-directional (just looking for any difference).
 Example:
You're testing if a new drug performs differently (could be better or worse):
- H_0: New drug performs the same as the old one.
- H_1: New drug performs differently than the old one.
- You care about any change, not just improvement.


8. What is the Z-test, and when is it used in hypothesis testing.

 - The Z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large (generally n \geq 30).

 What Is a Z-test?
- It compares a sample mean to a population mean—or compares two sample means.
- It uses the standard normal distribution (the Z-distribution).
- The test statistic follows a normal curve centered around 0.


9. How do you calculate the Z-score, and what does it represent in hypothesis testing.

- The Z-score is a standardized value that tells you how far a data point or sample mean is from the population mean, in terms of standard deviations. It's a cornerstone of the Z-test and helps determine if your result is statistically significant.

 Formula to Calculate Z-score
For a single sample mean:
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
Where:
- \bar{x}: Sample mean
- \mu: Population mean
- \sigma: Population standard deviation
- n: Sample size

- It shows how many standard deviations your sample mean is away from the population mean.
- The higher the absolute value of the Z-score, the more unusual the result under the null hypothesis.
- It helps you find the p-value, which is used to determine statistical significance.




10. What is the T-distribution, and when should it be used instead of the normal distribution.

- The T-distribution, also called Student's T-distribution, is a probability distribution that’s similar in shape to the normal distribution but has heavier tails. This means it accounts for more variability—especially in small sample sizes or when the population standard deviation is unknown.


- Bell-shaped and symmetric, like the normal curve.
- Has more spread (wider tails), which reflects increased uncertainty.
- As sample size grows, it becomes indistinguishable from the normal distribution.

14. What is a confidence interval, and how is it used to interpret statistical results.

- A confidence interval (CI) is a range of values that’s used to estimate an unknown population parameter—like the mean or proportion—with a certain level of confidence. Instead of giving just a single point estimate (e.g., sample mean), it gives a range that is likely to contain the true value.


15. What is the margin of error, and how does it affect the confidence interval.

- Margin of error (MoE) is the amount you add and subtract from a statistic (like a sample mean or proportion) to create a confidence interval. It reflects the uncertainty in your estimate based on sample variability.


- It’s the maximum expected difference between the true population value and your sample estimate.
- It comes from the critical value (like a Z or T-score) times the standard error.
 Formula:
\text{Margin of Error (MoE)} = z \cdot \frac{\sigma}{\sqrt{n}} \quad \text{or} \quad t \cdot \frac{s}{\sqrt{n}}
- z: Z-value (e.g., 1.96 for 95% confidence)
- t: T-value (for small samples or unknown standard deviation)
- \sigma: Population standard deviation
- s: Sample standard deviation
- n: Sample size




16. How is Bayes' Theorem used in statistics, and what is its significance.

- Bayes’ Theorem is a powerful tool in statistics that helps us update our beliefs based on new evidence. It’s at the heart of Bayesian statistics, which treats probability as a measure of belief or uncertainty—rather than just frequency.

 Bayes’ Theorem Formula
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
Where:
- P(A|B): Probability of event A given B (posterior)
- P(B|A): Probability of event B given A (likelihood)
- P(A): Prior probability of A
- P(B): Total probability of B

- Based on theory or prior data (e.g., each side of a fair die should have 1/6 probability)
- Apply the formula and compute the test statistic
- Determine degrees of freedom
\text{df} = k - 1- Where k is the number of categories
- Compare with critical value or find p-value

19. What is the F-distribution, and when is it used in hypothesis testing.

- The F-distribution is a continuous probability distribution that arises when comparing variances—most notably in ANOVA (Analysis of Variance) and regression analysis. It’s shaped like a skewed bell curve, typically skewed right, and is defined by two sets of degrees of freedom.


- It’s based on the ratio of two independent chi-square distributions divided by their respective degrees of freedom.
- Its values are non-negative, and its shape depends on two parameters:
- df_1: Degrees of freedom of the numerator
- df_2: Degrees of freedom of the denominator


20. What is an ANOVA test, and what are its assumptions.

- ANOVA (Analysis of Variance) is a powerful statistical test used to determine whether there are significant differences between the means of three or more groups. Instead of running multiple t-tests (which can increase the risk of errors), ANOVA lets you compare all group means in a single test.

 What ANOVA Tests
- Null Hypothesis (H_0): All group means are equal
- Alternative Hypothesis (H_1): At least one group mean is different
It's particularly handy when you're dealing with categorical independent variables and a continuous dependent variable.

22. What is the F-test, and how does it relate to hypothesis testing.

- The F-test is a statistical procedure used in hypothesis testing to compare variances between two or more groups. It's built on the F-distribution, and it's especially useful in models involving multiple variables or comparing multiple group means.


- It analyzes ratio of variances: typically the variance between groups vs. variance within groups.
- Produces an F-statistic:

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Returns the probability for a Bernoulli-distributed variable.

    Parameters:
    x (int): Outcome value (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: Probability of observing outcome x
    """
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return p if x == 1 else 1 - p


print("P(X = 1):", bernoulli_pdf(1, 0.7))
print("P(X = 0):", bernoulli_pdf(0, 0.7))
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(x, mean=0, std=1):
    z = calculate_z_score(x, mean, std)


    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, loc=mean, scale=std)


    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='orange', alpha=0.6, label=f'Z ≤ {z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Data Point Z = {z:.2f}')
    plt.title('Z-Score Visualization on Standard Normal Distribution')
    plt.xlabel('Z-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Data Point: {x}")
    print(f"Z-Score: {z:.2f}")

data_point = 85
population_mean = 70
population_std = 10

plot_z_score(data_point, population_mean, population_std)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm


np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=15, size=40)  # Mean shifted to 105 for testing


population_mean = 100
population_std = 15


sample_mean = np.mean(sample_data)
sample_size = len(sample_data)


z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))


p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats


data = [88, 92, 95, 91, 87, 90, 93, 89, 94, 96]


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)


confidence = 0.95
z_critical = stats.norm.ppf((1 + confidence) / 2)
margin_of_error = z_critical * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error


print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_hypothesis_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):

    se = sigma / np.sqrt(n)


    z_critical = norm.ppf(1 - alpha/2)
    margin = z_critical * se


    lower_crit = mu0 - margin
    upper_crit = mu0 + margin


    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    plt.figure(figsize=(12, 6))
    plt.plot(x, h0_pdf, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, h1_pdf, label='H₁: μ = {}'.format(mu1), color='green')


    plt.fill_between(x, h0_pdf, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.5, label='Type I Error (α)')


    plt.fill_between(x, h1_pdf, where=(x > lower_crit) & (x < upper_crit), color='orange', alpha=0.5, label='Type II Error (β)')


    plt.axvline(lower_crit, color='black', linestyle='--', label='Lower Critical Value')
    plt.axvline(upper_crit, color='black', linestyle='--', label='Upper Critical Value')

    plt.title('Visualizing Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()


visualize_hypothesis_errors()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data: list or numpy array of sample values
    - confidence_level: confidence level (e.g., 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)


    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z_score * se

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence_level * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)


sample_data = [88, 92, 85, 90, 91, 87, 89, 93, 94, 86]
confidence_interval(sample_data)
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import norm

def margin_of_error(data, confidence_level=0.95):

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)

    z_score = norm.ppf((1 + confidence_level) / 2)
    moe = z_score * se

    print(f"Sample Size: {n}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Confidence Level: {int(confidence_level * 100)}%")
    print(f"Z-score: {z_score:.2f}")
    print(f"Margin of Error: ±{moe:.2f}")

    return moe


sample_data = [88, 92, 85, 90, 91, 87, 89, 93, 94, 86]
margin_of_error(sample_data, confidence_level=0.95)
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_spam, likelihood_lottery_given_spam, likelihood_lottery_given_not_spam):

    prior_not_spam = 1 - prior_spam


    p_lottery = (likelihood_lottery_given_spam * prior_spam) + \
                (likelihood_lottery_given_not_spam * prior_not_spam)

    posterior_spam_given_lottery = (likelihood_lottery_given_spam * prior_spam) / p_lottery

    print(f"P(Spam|'lottery') = {posterior_spam_given_lottery:.4f}")
    return posterior_spam_given_lottery


bayes_theorem(prior_spam=0.4,
              likelihood_lottery_given_spam=0.9,
              likelihood_lottery_given_not_spam=0.1)
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import pandas as pd
import scipy.stats as stats

data = pd.DataFrame({
    'Age_Group': ['18-25', '18-25', '26-35', '26-35', '36-45', '36-45'],
    'Drink_Preference': ['Tea', 'Coffee', 'Tea', 'Coffee', 'Tea', 'Coffee'],
    'Count': [30, 20, 25, 35, 20, 40]
})


contingency_table = data.pivot(index='Age_Group', columns='Drink_Preference', values='Count')
print("Contingency Table:\n", contingency_table)

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)


print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", pd.DataFrame(expected,
                                              index=contingency_table.index,
                                              columns=contingency_table.columns))
```
